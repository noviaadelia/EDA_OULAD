```python
#Handling Missing Value
#Mengisis data yang kosong pada kolom date di tabel Assessment dengan nilai rata-ratanya
df_assessment['date'].fillna(df_assessment['date'].mean(), inplace=True)
```

```python
#Handling Missing Value
#Menghapus baris yang kosong pada kolom score di tabel Student Assessment
df_studentAss.dropna(subset=['score'], inplace=True)
```

```python
#Handling Missing Value
#Mengisi data yang kosong pada kolom imd_band di tabel Student Info dengan nilai modusnya
df_studentInfo['imd_band'].fillna(df_studentInfo['imd_band'].mode(), inplace=True)
```

```python
#Handling Missing Value
#Handling missig value pada kolom date_registration dan date_unregistration di tabel Student Registration
#dengan menghapus kolomnya karena data yang kosong terlalu banyak
del df_studentReg['date_registration']
del df_studentReg['date_unregistration']
```

```python
#Handling Missing Value
#Handling missig value pada kolom week_from dan week_to di tabel Vle
#dengan menghapus kolomnya karena data yang kosong terlalu banyak
del df_vle['week_from']
del df_vle['week_to']
```

**Merging Dataset**

Merging dataset adalah proses menggabungkan dua atau lebih dataset berdasarkan satu atau beberapa kolom kunci yang memiliki nilai yang sama di setiap dataset. Tujuan dari merge dataset adalah untuk menggabungkan informasi dari dataset yang berbeda menjadi satu dataset yang lebih lengkap atau terstruktur.

```python
#Penggabungan dataset df_studentInfo dan df_studentVle, berdasarkan kolom id_student dengan tipe merge "inner."
merged_df = df_studentInfo.merge(df_studentVle, on='id_student', how='inner')
```

```python
#Menampilkan 5 data teratas dari data yang telah digabungkan
merged_df.head()
```

**Visualisasi Data**

Bar Chart untuk menampilkan jumlah siswa berdasarkan jenis kelaminnya

```python
sns.countplot(data=merged_df, x='gender')
plt.title('Number of Students by Gender')
plt.xlabel('Gender')
plt.ylabel('Number of Students')
plt.show()
```

Pie Chart perbandingan hasil akhir yang diperoleh oleh siswa

```python
final_result_counts = merged_df['final_result'].value_counts()
plt.figure(figsize=(5, 5))
plt.pie(final_result_counts, labels=final_result_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Final Results')
plt.show()
```

Histogram untuk melihat distribusi kredit yang dipelajari oleh siswa

```python
plt.figure(figsize=(10, 6))
plt.hist(merged_df['studied_credits'], alpha=0.7)
plt.title('Distribution of Studied Credits')
plt.xlabel('Studied Credits')
plt.ylabel('Number of Students')
plt.grid(False)
plt.show()
```

Scatter Plot hubungan antara jumlah kredit yang dipelajari dan jumlah klik

```python
plt.figure(figsize=(10, 6))
plt.scatter(merged_df['studied_credits'], merged_df['sum_click'])
plt.title('Studied Credits & Sum Click')
plt.xlabel('Studied Credits')
plt.ylabel('Sum Click')
plt.grid(False)
plt.show()
```

Stacked Bar Chart perbandingan hasil akhir mahasiswa berdasarkan wilayah

```python
grouped_data = merged_df.groupby(['region', 'final_result']).size().unstack()
plt.figure(figsize=(12, 6))
grouped_data.plot(kind='bar', stacked=True)
plt.title('Number of Students per Region and Final Results')
plt.xlabel('Region')
plt.ylabel('Number of Students')
plt.xticks(rotation=90)
plt.legend(title='Final Result', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(False)
plt.show()
```

Heatmap korelasi untuk melihat hubungan antara variabel-variabel numerik berkorelasi satu sama lain.

```python
correlation_matrix = merged_df.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, linewidths=0.5)
plt.title('Heatmap Correlation')
plt.show()
```

Bar Chart tingkat pendidikan tertinggi dari siswa

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=merged_df, x='highest_education', palette='viridis')
plt.title('Highest Level of Education')
plt.xlabel('Education Level')
plt.ylabel('Number of Students')
plt.xticks(rotation=45)
plt.show()
```
